# Duplicati, tipi e outlier

Questo notebook esegue la lezione `duplicates-types-outliers` su un dataset sintetico.

```python
from pathlib import Path

import pandas as pd

from memory_ai.data_quality import clean_memory_quality_issues, duplicate_memory_mask

repo_root = Path('..')
raw_path = repo_root / 'datasets' / 'synthetic' / 'memory_events_quality_issues.csv'
raw = pd.read_csv(raw_path)
raw
```

```python
duplicate_mask = duplicate_memory_mask(raw)
raw.loc[duplicate_mask, ['memory_id', 'text', 'timestamp']]
```

```python
result = clean_memory_quality_issues(raw)
result.data
```

```python
result.report
```

```python
assert result.report['rows_before'] == 7
assert result.report['rows_after'] == 5
assert result.report['duplicates_removed'] == 2
assert result.report['invalid_importance_values'] == 1
assert result.report['importance_outliers'] == 2
assert result.data['importance'].between(0.0, 1.0).all()
```
