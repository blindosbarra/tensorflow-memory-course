# Duplicati, tipi e outlier

Questo notebook esegue la lezione su letture sintetiche di sensori ambientali.

```python
from pathlib import Path

import pandas as pd

repo_root = Path('..')
raw_path = repo_root / 'datasets' / 'synthetic' / 'environmental_sensor_quality_challenge.csv'
raw = pd.read_csv(raw_path)
raw
```

```python
duplicate_mask = raw.duplicated(['station_id', 'recorded_at'], keep='first')
raw.loc[duplicate_mask, ['reading_id', 'station_id', 'recorded_at']]
```

```python
numeric_temperature = pd.to_numeric(raw['temperature_c'], errors='coerce')
numeric_temperature.describe()
```

```python
numeric_temperature.loc[~numeric_temperature.between(-50, 60) | numeric_temperature.isna()]
```
